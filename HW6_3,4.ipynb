```python
#Your code goes here.
world_cup = pd.read_csv('/content/drive/MyDrive/DATA/WorldCupMatches.csv')
world_cup.head()
```

```python
# group and count
team_pairs = (
    world_cup
    .groupby(['Home Team Initials', 'Away Team Initials'])
    .count()
    .reset_index()
)

# keep only one count column and rename it to weight
team_pairs = team_pairs[['Home Team Initials', 'Away Team Initials', 'Year']]
team_pairs = team_pairs.rename(columns={'Year': 'weight'})

# preview
team_pairs.head()
```

```python
# total goals scored by each home team
home_goals = (
    world_cup
    .groupby('Home Team Initials')['Home Team Goals']
    .sum()
    .reset_index()
)

# rename column as required
home_goals = home_goals.rename(columns={'Home Team Goals': 'HomeGoalTotal'})

# preview
home_goals.head()
```

```python
team_pairs = team_pairs.merge(
    home_goals,
    on='Home Team Initials',
    how='left'
)

team_pairs.head()
```

```python
# create a sorted pair so ARG-BRA and BRA-ARG become the same
team_pairs['team_pair'] = team_pairs.apply(
    lambda row: tuple(sorted([row['Home Team Initials'], row['Away Team Initials']])),
    axis=1
)

# group and sum weights
team_pairs_combined = (
    team_pairs
    .groupby('team_pair', as_index=False)
    .agg({
        'weight': 'sum',
        'HomeGoalTotal': 'first'
    })
)

# split back into two columns
team_pairs_combined[['Home Team Initials', 'Away Team Initials']] = pd.DataFrame(
    team_pairs_combined['team_pair'].tolist(),
    index=team_pairs_combined.index
)

# drop helper column
team_pairs_combined = team_pairs_combined.drop(columns=['team_pair'])

team_pairs_combined.to_csv("team_pairs_combined.csv", index=False)
```
